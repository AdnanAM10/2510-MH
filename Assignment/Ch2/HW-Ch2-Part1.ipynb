```python
#Count the number of missing values in x1, x2 and x3  -> apply sum() method
missing_values_x1 = myData['x1'].isnull().sum()
missing_values_x2 = myData['x2'].isnull().sum()
missing_values_x3 = myData['x3'].isnull().sum()

# Print the number of missing values in x1, x2, and x3
print("Number of missing values in x1:", missing_values_x1)
print("Number of missing values in x2:", missing_values_x2)
print("Number of missing values in x3:", missing_values_x3)
```
