```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
exam_score = pd.DataFrame({
    "midterm": [65, 71, 72, 48, 55, 25, 55, 45, 37, 43, 58, 65, 57, 59, 9,100],
    "final":   [60, 68, 64, 55, 59, 45, 66,55, 38, 52, 54,34, 98, 43, 99,100]
})
```

```python
BoxPlot1 = exam_score.boxplot(column=['midterm', 'final'], vert=True)
BoxPlot1.set(ylabel='Scores', title='Exam Scores: Midterm vs Final')
plt.show()
```

```python
s = exam_score.dropna() #drop the rows with NaN (missing values) and store the result in s
```

```python
q1_midterm = s['midterm'].quantile(0.25)
med_midterm = s['midterm'].quantile(0.50)
q3_midterm  = s['midterm'].quantile(0.75)
IQR_midterm = q3_midterm - q1_midterm

lower_fence_midterm = q1_midterm - 1.5 * IQR_midterm
upper_fence_midterm = q3_midterm + 1.5 * IQR_midterm


print("min:", s['midterm'].min())
print("Q1:", q1_midterm)
print("median:", med_midterm)
print("Q3:", q3_midterm)
print("IQR:", IQR_midterm)

print("low_fence:", lower_fence_midterm) #Low fence value
print("High_fense:", upper_fence_midterm) #High fence value

```

```python
#midterm scores less than the upper_fence_midterm to find the whisker_high
s['midterm'] <= upper_fence_midterm #list of Trues/Falses

```

```python
# find the high whisker value for the midterm

midterm_less_than_upper_fence = s[s['midterm'] <= upper_fence_midterm]

# Among the midterm scores less than the midterm score upper fence, find the max value
whisker_high_midterm = midterm_less_than_upper_fence['midterm'].max()

print(whisker_high_midterm)
```

```python
# find the low whisker value for the midterm
midterm_higher_than_lower_fence = s[s['midterm'] >= lower_fence_midterm]
whisker_low_midterm = midterm_higher_than_lower_fence['midterm'].min()
print(whisker_low_midterm)
```

```python
#Find the outliers if any

# Any values greater than the higher fence

high_outlier = s[s['midterm'] >= upper_fence_midterm]
print(high_outlier)
print( "******************************")

low_outlier = s[s['midterm'] <= lower_fence_midterm]
print(low_outlier)
```
