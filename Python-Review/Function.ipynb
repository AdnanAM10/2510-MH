```python
def scale(width, height, /, factor=1.0):
    """width and height must be passed positionally; factor has a default."""
    return (width * factor, height * factor)

print(scale(100, 50))        # (100.0, 50.0)
print(scale(100, 50, 0.5))   # (50.0, 25.0)

#with and height are postional-only --> you must call like scale(100, 50, ...)
# The following is INVALID because width/height are positional-only:
# scale(width=100, height=50)  # TypeError
```

```python
#Default parameter value
#Default parameter value is the value a fucntion uses for a parameter when the caller does not supply one

def greet(name, punctuation="!"):
    return f"Hello, {name}{punctuation}"

print(greet("Ada"))         # Hello, Ada!
print(greet("Ada", "!!!"))  # Hello, Ada!!!
```

```python
#Integer is immutable!
x=5

def my_function (x):
  x=10
  print(x)

my_function(x)

print(x)
```

```python
# A list can be passed to a function
# Note that a list is mutable

x = [1,2,3,4]


def my_function (x):
  x[3] = 10
  print(x)

my_function(x)

print(x)

# Can you see the difference between the above and this code snippet?
```
