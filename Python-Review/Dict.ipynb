```python
#Access items:

car = {"brand":"Ford", "model":"Mustang", "year":1964}
x = car["model"]
display(x)
```

```python

car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.values()
y = car.keys()
z = car.items()

print(x)
print(y)
print(z, "\n")

# dictionary is mutable!

car["year"] = 2020  #change the values

print(car) #after the change

```

```python
# Add itmes!
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

car["color"] = "red"
car.update({"price":36000})
print(car)
```

## Remove Items


```python
# Add itmes!
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

car.pop("model")
display (car)
#car.popitem() #remove the last inserted item; before version 3.7, a random item is removed instead; not recommended
del car["year"]
print(car)
```

## Loop Dict

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

for x in car:
    print(x)

for x, y in car.items():
    print(x, y)

```

```python
#deep copy
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964,
"color": "blue"
}

another_car = car.copy()
another_car["color"] = "red"
display(another_car)

display(car)
print("\n")

#Shallow copy
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964,
"color": "blue"
}

another_car = car

another_car["color"] = "red"
display(another_car)

display(car)
```

## Nested Dict!

```python
child1 = {
  "name" : "Emil",
  "year" : 2004
} #dictionary


child2 = {
  "name" : "Tobias",
  "year" : 2007
} #dictionary


child3 = {
  "name" : "Linus",
  "year" : 2011
} # dictionary


# nested dictionary

myfamily = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
}
```
