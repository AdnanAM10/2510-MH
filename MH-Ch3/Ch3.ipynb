# Growth.quantile(0.30)

1. Take the numeric values in myData["Growth"] and ignore any NaNs (theyre dropped by default).

2. Sort the remaining values from smallest to largest.

3. Find the position (or index) p = (n-1) * 0.3, where n is the number of (non-NaN) values.  Note p stands for index (index starts from 0)

4.  p isnt an integer, interpolate linearly between the two surrounding values (this is the default behavior).



```python
#Example 3.4
#Import the Growth_Value data file and label it myData

import pandas as pd
myData = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/2510/2510-MH/MH-Ch3/jaggia_ba_2025_ch03_data.xlsx', sheet_name = 'Growth_Value')
myData

# Calculating measures of dispersion for Growth
print(myData.Growth.max() - myData.Growth.min())  # Range
print(myData.Growth.quantile(.75) - myData.Growth.quantile(.25)) #IQR
print((abs(myData.Growth-myData.Growth.mean())).mean()) #Mean Absoluate Deviation About the mean
print(myData.Growth.var()) #Sample variance ddof =1
print(myData.Growth.std()) #Sample std
print(myData.Growth.std()/myData.Growth.mean()) #CV

# Calculating measures of dispersion for Value
print(myData.Value.max() - myData.Value.min())
print(myData.Value.quantile(.75) - myData.Value.quantile(.25))
print((abs(myData.Value-myData.Value.mean())).mean())
print(myData.Value.var())
print(myData.Value.std())
print(myData.Value.std()/myData.Value.mean())
```
