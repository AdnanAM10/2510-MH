**The core idea in row subsetting: iloc and loc methonds**

.iloc → position-based (think: i = integer index/position)

*   You give row (position) numbers (0, 1, 2, …).
*   Slice end is exclusive.

.loc → label-based (think: l = label)

*   You give row labels (index values like dates, strings, IDs, or integers if your index is integers).
*   Slice end is inclusive.


If your index is somthing like: 0, 1,2,3..., be careful
*   df.iloc[0] --> first row by position  --> 0th position = 1st row
*   df.loc[0] --> the row whose label is 0

## What : means in slices [start:end]

start: where to begin (inclusive)

stop: where to end

.iloc (position-based): end is exclusive
e.g., df.iloc[0:3] → positions 0,1,2

.loc (label-based): end is inclusive
e.g., df.loc["A":"C"] → labels A, B, and C

**When to use axis = 0 or axis =1 --> The rule of thumb: **
```


axis=0 is the row/index axis (↓).

axis=1 is the column axis (→).

1) Reducers (sum/mean/min/max/…)

axis = the dimension you collapse over.

df.sum(axis=0) → sum down the rows → you get one value per column.

df.sum(axis=1) → sum across columns → you get one value per row.
This mirrors NumPy’s convention: “reduce along this axis.”

2) Label-based mutators (drop/sort/reindex/rename)

axis = which set of labels you’re editing/reordering.

df.drop(..., axis=0) → drop rows (edit the index labels).

df.drop(..., axis=1) → drop columns (edit the column labels).

df.sort_index(axis=0) → sort rows by index; axis=1 → columns by name.

df.sort_values(by=..., axis=0) → reorder rows by column values;
axis=1 → reorder columns by a row’s values (using by=<row label>).


```
