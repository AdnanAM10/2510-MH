**The core idea in row subsetting: iloc and loc methonds**

.iloc â†’ position-based (think: i = integer index/position)

*   You give row (position) numbers (0, 1, 2, â€¦).
*   Slice end is exclusive.

.loc â†’ label-based (think: l = label)

*   You give row labels (index values like dates, strings, IDs, or integers if your index is integers).
*   Slice end is inclusive.


If your index is somthing like: 0, 1,2,3..., be careful
*   df.iloc[0] --> first row by position  --> 0th position = 1st row
*   df.loc[0] --> the row whose label is 0

## What : means in slices [start:end]

start: where to begin (inclusive)

stop: where to end

.iloc (position-based): end is exclusive
e.g., df.iloc[0:3] â†’ positions 0,1,2

.loc (label-based): end is inclusive
e.g., df.loc["A":"C"] â†’ labels A, B, and C



```python
# Example code to show iloc vs loc
import pandas as pd
df = pd.DataFrame({
    "math": [85, 90, 78, 92],
    "english": [88, 76, 95, 80],
    "science": [91, 89, 84, 77],
    "history": [70, 85, 88, 93]
}, index=["s1", "s2", "s3", "s4"])

df
```
